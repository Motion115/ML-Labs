```python
import os
import pickle

# read
with open(os.path.join("./enrico/enrico_embedding_150.pkl"), 'rb') as f:
    corpus = pickle.load(f)

id2label = {}
# iterate the corpus
for i in range(len(corpus)):
    id2label[corpus[i]['id']] = corpus[i]['label_id']
print(id2label)
```

```python
import os
# load the image data
dir = "./enrico/image/"
# list all the files in the directory
files = os.listdir(dir)
# use PIL to load the image
from PIL import Image
# use numpy to convert the image pixels into a numpy array
import numpy as np

data = []
deprecated = ["50105.jpg", "50109.jpg"]
for file in files:
    if file in deprecated:
        continue 
    # load the image
    img = Image.open(dir + file)
    # resize the image to be 64x64 pixels
    img = img.resize((32, 32))
    # flatten the img into a 1d numpy array
    img = np.array(img).flatten()
    # split the filename to get the id
    id = file.split(".")[0]
    # get the label id from the id2label dictionary
    label_id = id2label[id]
    # append the image data and label id to the data list
    data.append({"id": id, "vector": img, "label_id": label_id})

# save the data
with open(os.path.join("./enrico/enrico_image_data.pkl"), 'wb') as f:
    pickle.dump(data, f)
```
